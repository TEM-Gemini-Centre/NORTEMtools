```python
%matplotlib qt5

# ðŸ”§ Imports
from pathlib import Path
import hyperspy.api as hs
import logging

# Path to the parent folder that contains multiple subfolders with .mib files
parent_dir = Path(r"D:/datasets/250225_Ag_oxww")  # <-- CHANGE THIS

# Options
recursive = True        # True = search all subfolders, False = only one level deep
overwrite = False       # True = overwrite existing .zspy files
dry_run = False         # True = only show actions without writing files
lazy = True             # True = use lazy loading (faster, less memory)
verbose = True          # Print detailed logs

# --- Setup logging ---
logging.basicConfig(level=logging.INFO if verbose else logging.WARNING,
                    format="%(levelname)s: %(message)s")
```

```python
def find_mib_files(parent: Path, recursive: bool = True):
    """
    Find .mib files once (no duplicates), case-insensitive, across platforms.
    Works on Windows/macOS/Linux. Avoids duplicate hits from multiple patterns.
    """
    if recursive:
        candidates = parent.rglob("*")
    else:
        candidates = parent.glob("*/*")

    unique = []
    seen = set()

    for p in candidates:
        if p.is_file() and p.suffix.lower() == ".mib":
            # Normalize the key to avoid duplicates on case-insensitive filesystems
            key = str(p.resolve()).casefold()
            if key not in seen:
                seen.add(key)
                unique.append(p)

    logging.info(f"Found {len(unique)} .mib files.")
    return sorted(unique)

mib_files = find_mib_files(parent_dir, recursive)
for f in mib_files:
    print(f)
```

```python
def convert_mib_to_zspy(mib_path: Path, overwrite=False, lazy=True):
    """Convert one .mib file to .zspy in the same directory."""
    out_path = mib_path.with_suffix(".zspy")
    
    if out_path.exists() and not overwrite:
        logging.info(f"Skipping (exists): {out_path}")
        return out_path

    logging.info(f"Loading: {mib_path}")
    s = hs.load(mib_path, lazy=lazy)

    logging.info(f"Saving as: {out_path}")
    s.save(out_path, overwrite=overwrite)
    return out_path
```

```python
converted = 0
skipped = 0

for mib_path in mib_files:
    out_path = mib_path.with_suffix(".zspy")
    
    if out_path.exists() and not overwrite:
        logging.info(f"Skipping existing: {out_path}")
        skipped += 1
        continue
    
    if dry_run:
        logging.info(f"[DRY-RUN] Would convert: {mib_path}")
        converted += 1
        continue
    
    try:
        convert_mib_to_zspy(mib_path, overwrite=overwrite, lazy=lazy)
        converted += 1
    except Exception as e:
        logging.error(f"âŒ Failed to convert {mib_path}: {e}")

print("\nâœ… Conversion Summary")
print("---------------------")
print(f"Total .mib files found : {len(mib_files)}")
print(f"Converted (incl. dry)  : {converted}")
print(f"Skipped (already exist): {skipped}")
```
