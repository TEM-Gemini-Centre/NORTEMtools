```python
%matplotlib qt5
import hyperspy.api as hs
import pyxem as pxm
import numpy as np
import orix
import matplotlib.pyplot as plt

from pathlib import Path
from diffpy.structure import Atom, Lattice, Structure
from diffsims.generators.simulation_generator import SimulationGenerator
```

```python
file_path = Path("D:/datasets/mini.hspy")
file_path2 = Path("D:/datasets/cropped_Ag.hspy")
file_path3 = Path("D:/datasets/SPED_Ag_full.hspy")

dp_mini = hs.load(file_path, lazy=True)
dp_cropped = hs.load(file_path2, lazy=True)
dp = hs.load(file_path3, lazy=True)
```

```python
dp_mini.axes_manager[2].scale = 0.02741/2
dp_mini.axes_manager[3].scale = 0.02741/2

dp_mini.axes_manager[2].units = "A^-1"
dp_mini.axes_manager[3].units = "A^-1"

dp_cropped.axes_manager[2].scale = 0.02741/2
dp_cropped.axes_manager[3].scale = 0.02741/2

dp_cropped.axes_manager[2].units = "A^-1"
dp_cropped.axes_manager[3].units = "A^-1"

dp.axes_manager[2].scale = 0.02741/2
dp.axes_manager[3].scale = 0.02741/2

dp.axes_manager[2].units = "A^-1"
dp.axes_manager[3].units = "A^-1"
```

```python
s_shifts = dp_mini.get_direct_beam_position(method="blur", sigma=5, half_square_width=20)
s_shifts2 = dp_cropped.get_direct_beam_position(method="blur", sigma=5, half_square_width=20)
s_shifts3 = dp.get_direct_beam_position(method="blur", sigma=5, half_square_width=20)

s_shifts.compute()
s_shifts2.compute()
s_shifts3.compute()

s_linear_plane = s_shifts.get_linear_plane()
s_linear_plane2 = s_shifts2.get_linear_plane()
s_linear_plane3 = s_shifts3.get_linear_plane()
```

```python
dp_mini_centered = dp_mini.center_direct_beam(shifts=s_linear_plane, inplace=False)
dp_cropped_centered = dp_cropped.center_direct_beam(shifts=s_linear_plane2, inplace=False)
dp_centered = dp.center_direct_beam(shifts=s_linear_plane3, inplace=False)

dp_mini_centered.save('mini_centered.hspy')
dp_cropped_centered.save('cropped_Ag_centered.hspy')
dp_centered.save('SPED_Ag_full_centered.hspy')
```

```python
dp_template = dp_mini_centered.template_match_disk(disk_r = 3, subtract_min=False)
```
